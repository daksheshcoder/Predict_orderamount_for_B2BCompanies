```python
def box_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    mask = (series < Q1 - 1.5 * IQR) | (series > Q3 + 1.5 * IQR)
    series = series[mask]
    return series
```

```python
Q1 = df['amount_in_usd'].quantile(0.25)
Q3 = df['amount_in_usd'].quantile(0.75)
IQR = Q3 - Q1

outlier_threshold = 1.5

median_value = df['amount_in_usd'].median()
outliers = df[(df['amount_in_usd'] < Q1 - outlier_threshold * IQR) | (df['amount_in_usd'] > Q3 + outlier_threshold * IQR)]
df.loc[outliers.index, 'amount_in_usd'] = median_value

total_outliers = len(outliers)

print("DataFrame with outliers replaced:")
print(df)
print("\nTotal number of outliers:", total_outliers)

```

# Linear Regression

```python

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score


selected_columns = ['SALES_ORG','COMPANY_CODE','CREDIT_STATUS','CUSTOMER_NUMBER','unique_cust_id','DISTRIBUTION_CHANNEL_encoded',
'DIVISION_encoded','ORDER_AMOUNT_LOG']
X = df[selected_columns]
y = df['amount_in_usd_log']

# Convert YearMonth to year and month columns
X['Year'] = df['YearMonth'].dt.year
X['Month'] = df['YearMonth'].dt.month

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print(y_pred)
```

 Coverting the log normalized values to Normal form

```python
predictions_log = model.predict(X_test) 


predictions = np.exp(predictions_log)


print(predictions)
```

```python
y_pred = model.predict(X_test)


mse = mean_squared_error(y_test, y_pred)


rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
print("Mean Absolute Error (MAE):", mae)
print("R-squared (R^2):", r2)
```

 Checking the overfitting or underfitting of the model

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.model_selection import RandomizedSearchCV

selected_columns = ['SALES_ORG', 'COMPANY_CODE', 'CREDIT_STATUS', 'CUSTOMER_NUMBER','ORDER_AMOUNT_LOG']
X = df[selected_columns]
y = df['amount_in_usd_log']

# Convert YearMonth to year and month columns
X['Year'] = df['YearMonth'].dt.year
X['Month'] = df['YearMonth'].dt.month

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Define the AdaBoost regression model
model = AdaBoostRegressor()

# Define the parameter grid for RandomizedSearchCV
param_grid = {
    'n_estimators': [50, 100, 150],
    'learning_rate': [0.01, 0.1, 1.0],
    'loss': ['linear', 'square', 'exponential']
}

# Perform Randomized Search Cross Validation
random_search = RandomizedSearchCV(estimator=model, param_distributions=param_grid, n_iter=2, cv=5, random_state=42)
random_search.fit(X_train, y_train)

# Get the best model and its parameters
best_model = random_search.best_estimator_
best_params = random_search.best_params_

# Make predictions on the test set
y_pred = best_model.predict(X_test)

# Calculate evaluation metrics
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Print the evaluation metrics
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
print("Mean Absolute Error (MAE):", mae)
print("R-squared (R^2):", r2)
```
